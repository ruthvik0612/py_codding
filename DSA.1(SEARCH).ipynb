## Brute force

```python
# the number of cards is sorted in order , 
#now we have to find the user given number
# in the fastest way or path 
```

```python
#Create a variable position with the value 0.
#Check whether the number at index position in card equals query.
#If it does, position is the answer and can be returned from the function
#If not, increment the value of position by 1, and repeat steps 2 to 5 till we reach the last position.
#If the number was not found, return -1.
```

```python
def locate_card(cards,query):
    #create a variable position with a value 0 
    position = 0
    #set a while for repetation
    while True:
        #if the element at current position matches the query
        if cards[position] == query:
            #IF YES RETURN POSITION AND EXIT
                return position
        #if not increment the postion
        position +=1 

        if position == len(cards):
             #if the array is end
             return -1 # then returns -1


```

```python
test = {
    'input': { 
        'cards': [13, 11, 10, 7, 4, 3, 1, 0], 
        'query': 7
    },
    'output': 3
}

locate_card(**test['input']) == test['output']
```

```python
def locate_card(cards,query):
    position = 0
    while position < len(cards): # updated the query position < len(cards)
        if cards[position] == query:
            return position
        position +=1 
    return -1 


```
