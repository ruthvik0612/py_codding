Basic functions

```python
def add(x,y):
    c = x+y
    return c
add(10,20)
```

```python
#to find the factorial 

def fact(num):
    if num == 1 or num == 0:
        print(1)
    else:
        fact = 1
        for i in range(1,num+1):
            fact = fact*i

        
        print(fact)
fact(12)
```

```python
#we can add 2 number what if i want to add n numbers given by user

def add(*num):
    add = 0
    for i in num:
        add = add+i
    return add

add(12,12)
```

## lamba

```python

def multiply(a,b,c):
    x = lambda a,b,c:a*b*c
    return x(a,b,c)
multiply(1,1,2)
```

```python
product = lambda a,b,c:a*b*c
print(product(1,2,3))
```
