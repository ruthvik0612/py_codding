```python
#types of file 
# text files --> txt,docx,...
# binary files --> png,mp3.mp4,mov,....

## open ,read,write 
#open("filename", "mode")
#"r" → read  
#"w" → write (overwrites file)   clear all data from exisitng data
#"a" → append (adds data)    # add new data to existing data
```

```python
#read the file 
with open("rc.txt","r") as f :
    data = f.read()
    print(data)
```

```python
with open("rc.txt","w") as f:
    f.write("i want to learn data science ")
```

```python
with open("rc.txt","a") as f:
    f.write("\n i am ruthvik")
with open("rc.txt","r") as f:
    data = f.read()
    print(data)
```

```python
with open("rc.txt","r") as f:
    line1 = f.readline()
    line2 = f.readline()
    print(line1)
    print(line2)
```
