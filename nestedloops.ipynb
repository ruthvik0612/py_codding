# we follow star patterns to learn nested loops

```python
# * * * * *
# * * * * *
# * * * * *
# * * * * *
# * * * * *

for i in range(1,6):
    for j in range(1,6):
        print("*", end = " ")
    print()
```

```python
for i in range(1,6):
    print(i*"*")
```

```python
r = 6
for i in range(1,r+5):
    for j in range (1,i+1):
        print("*",end = "")
    print()
```
