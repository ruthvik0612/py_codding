## INHERITNACE

```python
class ben10:
    def __init__(self,name):
        self.name = name
       
    def intro(self):
        print(f"hello i am {self.name}")
class XLR8(ben10):
    def __init__(self,name,power):
        super().__init__(name) #to get name from parent class
        self.power = power
    def power_details(self):
        print(f"my power is to move {self.power}")

#multilevel inheritance

class fourarms(ben10):
    def __init__(self, name,power):
        super().__init__(name)
        self.power = power
    def power_details(self):
        print(f"my power is to move {self.power}")


    
```

```python
xlr8 = XLR8("BEN10" ,"FASTER")
xlr8.power_details()
```
