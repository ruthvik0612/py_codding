```python
#task 3
orders = []
while True:
    x =input("enter the value \n 1 to add amount to list \n 2 to show all orders \n q to quit")
    if x == "1":
        i = int(input(" enter i value"))
        orders.append(i) 
    elif x == "2":
        if len(orders) == 0:
            print("empty list")
        else:
            for items  in orders:
                print(items)
                
    elif x == "q":
        print("exiting")
        break
    else: 
        print("In valid input")
        continue
        
    
```
