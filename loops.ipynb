```python
#print the elements of a list
lst= ['a',12,'ruthvik','chowdary']

for i  in lst :
    print(i)
```

```python
# if i want retrevie the each char of str
# is instance learnt from gtp
lst= ['a',12,'ruthvik','chowdary']

for i  in lst :
    if isinstance(i,str):
        for ch in i:
            print(ch, end = "-")
    else:
        print(i)
        
```
