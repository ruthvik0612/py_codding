## **mini sales analysis** on numpy 

```python
sales = np.array([1200,1500,900,2000,1800,1700,1600])
#total weekly sales
np.sum(sales)
```

```python
#avg daily sales
avg_sales = np.average(sales)
avg_sales
```

```python
# highest sales 
np.max(sales)
```

```python
#lowest of sales
np.min(sales)
```

```python
#standerd deviation
np.std(sales)
```

```python
#sales above the average
sales[sales > avg_sales]
```
