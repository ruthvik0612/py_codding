```python
#task 1
def apply_discount(price,discount_percent=5):

    final_price = price-((discount_percent/100)*price)
    return final_price

apply_discount(1000,10)
```

```python
#task 2 
def factorial(n):
    if n < 0 :
        print("its a -ve number")
        return

    elif n == 0 or n == 1:
        return 1
    return n*factorial(n-1)

print(factorial(5))
print(factorial(0))
factorial(-1)

```

```python
#task 3
def get_price(price):
    gst = lambda  price : price + (0.8 * price)
    return gst(price) # calling lambda
get_price(100)
```

```python
# task 6
prices =[100,500,900,50,750]
def process_prices(prices):
    discounted_prices  = list(map(lambda prices : prices+ (prices * 0.10),prices))
    dis_price_300  = list(filter(lambda discounted_prices : discounted_prices > 300 ,discounted_prices))
    return discounted_prices,dis_price_300
    
process_prices(prices)
```

```python

#mini problem

price_list = []
def add_price():
    x = int(input("enter a price"))
    price_list.append(x)

def avg_price():
    if len(price_list) == 0:
        print("price_list is empty")
    else:
        avg_list= sum(price_list)/len(price_list)
        print(f"The avg price is {avg_list}")
    
def max_price():
    if len(price_list) == 0:
        print("price_list is empty")
    else:
        max_list = max(price_list)
        print(f" The max price from list is {max_list}")

while True:
    choice = input("1 add | 2 avg | 3 max | q quit: ")

    if choice == "1":
        add_price()
    elif choice == "2":
        avg_price()
    elif choice == "3":
        max_price()
    elif choice == "q":
        break
    
```
